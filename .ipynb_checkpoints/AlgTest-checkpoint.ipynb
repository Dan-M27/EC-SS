```python
# Takes an array of vectors as the argument
def test_f(x):
    x1 = x[:,0]
    x2 = x[:,1]
    return -(( 4 - 2.1*np.power(x1,2) + np.power(x1,4)/3 ) * np.power(x1,2) + x1*x2 + (
        -4 + 4*np.power(x2,2)) * np.power(x2,2))
    
```
